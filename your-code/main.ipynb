A and B are events of a probability space with $(\omega, \sigma, P)$ such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are false?
* $P(A \cup B) = 0.6$
* $P(A \cap B^{C}) = 0.2$
* $P(A \cap (B \cup B^{C})) = 0.4$
* $P(A^{C} \cap B^{C}) = 0.3$
* $P((A \cap B)^{C}) = 0.9$

```python
"""
1) P(A ∪ B) = 0.6?

P(A) = 0.3
P(B) = 0.6
P(A ∩ B) = 0.1

P(A ∪ B) = P(A) + P(B) - P(A ∩ B)
=> 0.6 = 0.3 + 0.6 - 0.1
=> 0.6 = 0.9 - 0.1
=> 0.6 != 0.8

Therefore the 1) is false.

2) P(A ∩ Bc) = 0.2?

P(Bc) = 1 - P(B)
=> P(Bc) = 1 - 0.6
=> P(Bc) = 0.4

&
P(A ∩ Bc) = P(A) * P(Bc)
=> P(A ∩ Bc) = 0.3 * 0.6
= > P(A ∩ Bc) = 0.18 != 0.2 

Therefore the statement is false.

3) P(A ∩ (B ∪ Bc)) = 0.4?

P(B ∪ Bc) = P(B) + P(Bc) - P(B ∩ Bc)
=> P(B ∪ Bc) = 0.6 + 0.4 - 0
=> P(B ∪ Bc) = 1.0

&

P(A ∩ (B ∪ Bc)) = P(A) * P(B ∪ Bc)
=> P(A ∩ (B ∪ Bc)) = P(A) * 1
=> P(A ∩ (B ∪ Bc)) = P(A)
=> P(A ∩ (B ∪ Bc)) = 0.3

0.3 != 0.4

Therefore, the statement is False.

4) P(Ac ∩ Bc) = 0.3?
P(Ac) = 1 - P(A)
=> P(Ac) = 1 - 0.3
=> P(Ac) = 0.7

&

P(Ac ∩ Bc) = P(Ac) * P(Bc)
=> P(Ac ∩ Bc) = 0.7 * 0.4
=> P(Ac ∩ Bc) = 0.28

therefore the statement is False.

5) P((A ∩ B)c) = 1 - P(A ∩ B)
P((A ∩ B)c) = 1 - 0.1
P((A ∩ B)c) = 0.9

0.9 == 0.9
Therefore the statement is True.
"""
```
